#### 1. For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one. *NOTE*: For illustration, only one of the three dates is shown in the starter code below.

```python
# For the date early in the dataset, measure the arbitrage spread between the two exchanges
# by subtracting the lower-priced exchange from the higher-priced one

arbitrage_spread_early = pd.DataFrame(data=(coinbase_early - bitstamp_early))
arbitrage_spread_early.drop_duplicates()
arbitrage_spread_early.dropna()


arbitrage_spread_mid = pd.DataFrame(data=(coinbase_mid - bitstamp_mid))
arbitrage_spread_mid.drop_duplicates()
arbitrage_spread_mid.dropna()



arbitrage_spread_late = pd.DataFrame(data=(coinbase_late - bitstamp_late))
arbitrage_spread_late.drop_duplicates()
arbitrage_spread_late.dropna()

# Use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame
# YOUR CODE HERE
profitable_spread_early = arbitrage_spread_early[arbitrage_spread_early > 0].dropna()
profitable_spread_middle = arbitrage_spread_middle[arbitrage_spread_middle > 0].dropna()
profitable_spread_late = arbitrage_spread_late[arbitrage_spread_late > 0].dropna()

profitable_spread_early.describe()
```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).
spread_return_early = pd.DataFrame(arbitrage_spread_early[arbitrage_spread_early > 0]['Close'] / bitstamp_early)
spread_return_mid = pd.DataFrame(arbitrage_spread_mid[arbitrage_spread_mid > 0]['Close'] / bitstamp_mid)
spread_return_late = pd.DataFrame(arbitrage_spread_late[arbitrage_spread_late > 0]['Close'] / bitstamp_late)
# Review the spread return DataFrame
# YOUR CODE HERE
spread_return_early.head()
```

#### 3. For each of the three dates, narrow down your trading opportunities even further. To do so, determine the number of times your trades with positive returns exceed the 1% minimum threshold that you need to cover your costs.

```python
# For the date early in the dataset, determine the number of times your trades with positive returns 
# exceed the 1% minimum threshold (.01) that you need to cover your costs
profitable_trade_early = pd.DataFrame(spread_return_early[spread_return_early > .01]['Close'].dropna())
profitable_trade_mid = pd.DataFrame(spread_return_mid[spread_return_mid > .01]['Close'].dropna())
profitable_trade_late = pd.DataFrame(spread_return_early[spread_return_early > .01]['Close'].dropna())


# Review the first five profitable trades
# YOUR CODE HERE
profitable_trades = pd.concat([profitable_trade_early,profitable_trade_mid,profitable_trade_late])
profitable_trades.head()
```
